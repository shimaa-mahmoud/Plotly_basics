### Subplots

```python
df = px.data.iris()
 # barmode : 'group'`, `'overlay'` or `'relative'
fig = px.bar(df, x="sepal_width", y="sepal_length",
             color="species" # different species have diff colors
             , barmode="group", # stack bar next to eachother
             facet_row="species" # each row represent species  (make it 3*1 plots)
             , facet_col="species_id") # each col represent species id (make it 3*3 plots)
 
fig.show('png')
```

```python
df = px.data.iris()
 
fig = px.bar(df, x="sepal_width", y="sepal_length",
             color="species", barmode = 'group')
fig.show('png')
```

```python
df = px.data.iris()
 
fig = px.bar(df, x="sepal_width", y="sepal_length",
             color="species", barmode = 'overlay')
fig.show('png')
```

## 3.Histograms

```python
help(px.histogram)
```

```python
df = px.data.tips()
  
fig = px.histogram(df, x="total_bill")
fig.show('png')
```

### Types of normalizaion
1. **percent or probability**: The output of histfunc for a given bin is divided by the sum of the output of histfunc for all bins.`histnorm = percent`
2. **density**: The output of histfunc for a given bin is divided by the size of the bin.`histnorm = density`
3. **probability density**:The output of histfunc for a given bin is normalized such that it corresponds to the probability that a random `histnorm = probability density`

```python
df = px.data.tips()
  
fig = px.histogram(df, x="total_bill", histnorm = 'percent')
fig.show('png')
```

```python
# you can stack histograms with color as the hue
df = px.data.tips()
  
fig = px.histogram(df, x="total_bill", color = "smoker")
fig.show('png')
```

**`marginal`: One of `'rug'`, `'box'`, `'violin'`, or `'histogram'`. If set, a
        subplot is drawn alongside the main plot, visualizing the distribution.**

```python
df = px.data.tips()
  
fig = px.histogram(df, x="total_bill", marginal = 'violin')
fig.show('png')
```

## 7.Gantt Chart


```python

import plotly.figure_factory as ff
  
df = [dict(Task="A", Start='2020-01-01', Finish='2009-02-02'),
      dict(Task="Job B", Start='2020-03-01', Finish='2020-11-11'),
      dict(Task="Job C", Start='2020-08-06', Finish='2020-09-21')]
  
fig = ff.create_gantt(df)
fig.show('png')
```

```python
df = [dict(Task="A", Start='2020-01-01', Finish='2009-02-02'),
      dict(Task="Job B", Start='2020-03-01', Finish='2020-11-11'),
      dict(Task="Job C", Start='2020-08-06', Finish='2020-09-21')]
  
colors = ['#7a0504', (0.2, 0.7, 0.3), 'rgb(210, 60, 180)']
  
fig = ff.create_gantt(df, show_colorbar = True, colors = colors    )
fig.show('png')
```
